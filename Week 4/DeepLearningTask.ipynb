```python
# Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import h2o
h2o.init()
```

```python
# Import data
url = "http://coursera.h2o.ai/cacao.882.csv"
data = h2o.import_file(url)
train, test = data.split_frame([0.8], seed = 50)
```

```python
# Set the dependent and explanatory variables
y = 'Maker Location'
ignoreFields = ['Maker Location']
x = [i for i in data.names if i not in ignoreFields]
```

```python
# Baseline model (1'41'' on my laptop)
from h2o.estimators.deeplearning import H2ODeepLearningEstimator
baseline = H2ODeepLearningEstimator(seed = 10, nfolds = 6) 
%time baseline.train(x, y, train)
```

```python
baseline.model_performance(train) # l1 = 0.00001
```

```python
baseline.model_performance(test)
```

```python
from h2o.grid.grid_search import H2OGridSearch

grid = H2OGridSearch(H2ODeepLearningEstimator(seed = 10, nfolds = 6), 
                            hyper_params = {
                                'activation': ["RectifierWithDropout"],
                                'rate' : [0.005, 0.001, 0.01],
                                'l1': [0, 0.00001, 0.0001],
                                'l2': [0, 0.00001, 0.0001], 
                                'input_dropout_ratio' : [0, 0.1],
                                'hidden_dropout_ratios':[[0, 0], [0.2,0.2]]
                            })

%time grid.train(x, y, train)
```

```python
grid.model_performance(test)
gbm_gridperf1 = grid.get_grid(sort_by='mse', decreasing=True)
gbm_gridperf1

# Grab the top GBM model, chosen by validation AUC
#best_gbm1 = gbm_gridperf1.models[0]

# Now let's evaluate the model performance on a test set
# so we get an honest estimate of top model performance
#best_gbm_perf1 = best_gbm1.model_performance(test)
```
